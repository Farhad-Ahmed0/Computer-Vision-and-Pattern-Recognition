The step activation function is a binary function that produces either 0 or 1 depending on the input. It is simple and efficient, making it useful in binary classification problems. However, it has limitations as it cannot be used in more complex tasks or multi-class classification. The main problem with the step function is that it is not differentiable, which makes it unsuitable for backpropagation-based learning algorithms. As a result, it loses any knowledge of gradients in input values and ignores any input value just below the threshold. More advanced activation functions such as sigmoid, ReLU, and tanh are commonly used in deep learning neural networks to overcome these limitations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step Function
def step(x):
    return np.heaviside(x, 1)
x = np.linspace(-10, 8, 1000)
y = step(x)
plt.plot(x, y)
plt.title("Step Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

The Sigmoid function is a non-linear function that outputs values between 0 and 1, making it suitable for both classification and regression tasks. It can provide a smooth transition from 0 to 1 and can return the probability output for each class in multiclass classification. However, it suffers from the vanishing gradient problem, and it is not a zero-centered function, which can affect the performance of some machine learning algorithms.Additionally, it is computationally expensive and has an exponential nature.

```python
# Sigmoid Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
x = np.linspace(-15, 8, 1000)
y = sigmoid(x)
plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

The Tanh activation function is a continuous function that is similar to the Sigmoid function, but it is symmetric over the origin and returns values between -1 and 1. It is commonly used in classification problems and provides zero-centered outputs, which solves the problem of all values being of the same sign. Additionally, it is non-linear, making it easy to backpropagate errors. However, like the Sigmoid function, it suffers from the vanishing gradient problem and is computationally expensive due to its exponential nature.

```python
# Tanh Function
def tanh(x):
    return np.tanh(x)
x = np.linspace(-15, 8, 100)
y = tanh(x)
plt.plot(x, y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Hyperbolic Tangent (tanh) Function')
plt.show()
```

The ReLU (rectified linear unit) activation function returns 0 or the input value, depending on whether the input is greater than or less than 0. It is commonly used in deep learning networks, and it is simple to backpropagate errors due to its non-linearity. ReLU accelerates the convergence of stochastic gradient descent, making it computationally efficient and capable of outputting a true zero value. However, its output is not zero-centered, which can decrease the neural network's efficiency during backpropagation. The dying ReLU problem can occur when the neuron gets stuck on the negative side and constantly outputs zero, either due to a high learning rate or significant negative bias.

```python
# Relu Function
def relu(x):
    return np.maximum(0, x)
x = np.linspace(-15, 8, 1000)
y = relu(x)
plt.plot(x, y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Rectified Linear Unit (ReLU) Function')
plt.show()
```

The ELU (Exponential Linear Unit) function is similar to ReLU, but it outputs negative values for negative inputs to ensure active neurons. It does not suffer from the problem of vanishing or exploding gradients or dying neurons. ELU leads to lower training time and higher accuracy in neural networks than ReLU and its variants. It is continuous and differentiable at all points, pushes the mean of activations closer to zero, and causes faster learning and convergence. However, ELU is slower to compute due to its non-linearity with negative inputs.

```python
# Elu Function
def elu(x, alpha=1.0):

    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
x = np.linspace(-15, 8, 1000)
elu1 = elu(x, alpha=1.0)
plt.plot(x, elu1, label='alpha = 1.0')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('ELU Activation Function')
plt.legend()
plt.show()
```

SELUs (Scaled Exponential Linear Units) are activation functions that induce self-normalization, resulting in neuronal activations with a zero mean and unit variance. Similar to ELUs, SELUs are scaled to ensure that the output of the entire layer remains constant, which reduces the problem of vanishing gradients. Compared to ReLUs, SELUs cannot die, learn faster and better, and do not need further processing. However, SELUs are a relatively new activation function and are not yet widely used in practice.

```python
# Selu Function
def selu(x, alpha=1.67326, scale=1.0507):
    positive = scale * (np.maximum(x, 0))
    negative = scale * (alpha * (np.exp(np.minimum(x, 0)) - 1))
    return positive + negative
x = np.linspace(-15, 8, 1000)
y = selu(x)
plt.plot(x, y)
plt.title("SELU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
